```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
def Funcion(Imagen, nFiltros=3, tFiltro=3, maxIters=2):
    #Creamos los filtros
    Filtros = []
    #Hola xd
    for i in range(nFiltros):
        filtro = np.random.uniform(-1, 1, (tFiltro, tFiltro))
        Filtros.append(filtro)

    #Sacamos la info de la imagen: Lenna es cuadrada
    filas, columnas = Imagen.shape
    
    #Aplicamos los filtros
    Salidas = [Imagen]*nFiltros

    for iteracion in range(maxIters):
        tamSal = int(((Salidas[0].shape[0] - 3)/ 1) + 1)
        for filtro in range(nFiltros):
            Salida = np.zeros((tamSal, tamSal))
            for fila in range(tamSal):
                for columna in range(tamSal):
                    # Aplicamos el filtro y luego ReLU
                    pixel = np.sum(Salidas[filtro][fila:fila + tFiltro, columna:columna + tFiltro] * Filtros[filtro])
                    Salida[fila, columna] = max(0, pixel)
            Salidas[filtro] = Salida

    # Calcular la imagen promedio de las salidas
    SalidaFinal = np.mean(Salidas, axis=0)
    
    return SalidaFinal
```

```python
lenna = cv.imread('lenna.png', 0)
sal = Funcion(lenna)
```

```python
plt.imshow(sal, cmap = 'gray')
```
